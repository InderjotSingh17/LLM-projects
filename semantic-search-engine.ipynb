```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
documents = [
    "Artificial Intelligence is transforming healthcare.",
    "Football is a popular sport worldwide.",
    "Deep learning is a subset of machine learning.",
    "Pizza is one of the most loved foods.",
    "Neural networks are used in AI."
]
```

```python
document_embeddings = model.encode(documents)
```

```python
print(document_embeddings.shape)
```

```python
query = "How can I learn artificial intelligence?"
```

```python
query_embedding = model.encode([query])
```

```python
similarities=cosine_similarity(
    query_embedding,
    document_embeddings
)
print(similarities)
```

```python
best_index=similarities.argmax()
print(best_index)
```

```python
for doc, score in zip(documents, similarities[0]):
    print(f"{score:.4f} --> {doc}")
```

```python
import faiss
import numpy as np
```

```python
document_embeddings=np.array(document_embeddings).astype("float32")
```

```python
dimension = document_embeddings.shape[1]
```

```python
index=faiss.IndexFlatL2(dimension)
```

```python
index.add(document_embeddings)
```

```python
print(index.ntotal)
```

```python
query = "How do I learn Artificial Intelligence?"
query_embedding = model.encode([query])
query_embedding = np.array(query_embedding).astype("float32")
```

```python
distances, indices = index.search(query_embedding, k=3)
```

```python
print(indices)
```

```python
for idx in indices[0]:
    print(documents[idx])
```
